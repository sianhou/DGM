```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.stats import multivariate_normal

# 定义两个峰
mu1 = [0, 0]
sigma1 = [[0.4, 0], [0, 0.4]]

mu2 = [1, 1]
sigma2 = [[0.5, 0], [0, 0.5]]

rv1 = multivariate_normal(mean=mu1, cov=sigma1)
rv2 = multivariate_normal(mean=mu2, cov=sigma2)

# 网格
x = np.linspace(-3, 4, 200)
y = np.linspace(-3, 4, 200)
X, Y = np.meshgrid(x, y)
pos = np.dstack((X, Y))

# 双峰联合 PDF
P = 0.5 * rv1.pdf(pos) + 0.5 * rv2.pdf(pos)

# 可视化
plt.figure(figsize=(6, 5))
plt.contourf(X, Y, P, levels=50, cmap='viridis')
plt.colorbar(label='Density')
plt.xlabel('X')
plt.ylabel('Y')
plt.title('2D Bimodal Joint PDF')
plt.show()
```
